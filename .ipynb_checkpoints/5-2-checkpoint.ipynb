```python
import os
import glob
import random
import shutil
from icrawler.builtin import BingImageCrawler
```

```python
# 画像を保存するルートディレクトリパス
root_dir = "fishimages/"
# 魚画像検索キーワードリスト
fishnames = ["アイゴ", "オニカサゴ", "カサゴ", "カワハギ", "キュウセンベラ", "クサフグ",
             "ソウシハギ", "マハゼ", "マアジ", "マイワシ", "ミノカサゴ", "メジナ", "メバル"]
# 収集画像データ数
data_count = 100
```

```python
def crawl_image(fishname, datacount, root_dir):
    
    crawler = BingImageCrawler(storage={"root_dir":root_dir + fishname + "/train"})
    crawler.session.verify = False
    
    filters = dict(
                    size="large",
                    type="photo"
                   )
    
    # クローリングの実行
    crawler.crawl(
                    keyword=fishname,
                    filters=filters,
                    max_num=datacount
                  )
    
    # 前回実行時のtestディレクトリが存在する場合、ファイルをすべて削除する
    if os.path.isdir(root_dir + fishname + "/test"):
        shutil.rmtree(root_dir + fishname + "/test")
    os.makedirs(root_dir + fishname + "/test")
    
    # ダウンロードファイルのリストを取得
    filelist = glob.glob(root_dir + fishname + "/train/*")
    # ダウンロード数の2割をtestデータとして抽出
    test_ratio = 0.2
    testfiles = random.sample(filelist, int(len(filelist) * test_ratio))
    
    for testfile in testfiles:
        shutil.move(testfile, root_dir + fishname + "/test/")
```

```python
for fishname in fishnames:
    crawl_image(fishname, data_count, root_dir)
```

```python
import glob
from azure.cognitiveservices.vision.customvision.prediction import CustomVisionPredictionClient
```

```python
base_url = 'https://fish-test.cognitiveservices.azure.com/'
project_ID = 'edc2988c-defc-4e5d-b9cc-04b4fe6d9f68'
publish_iteration_name = 'test1'
prediction_key = 'c35925bcd3af4e4cac754fe34a63579a'
```

```python
def testModel(testfiles, fishname):
    data_count = len(testfiles)
    true_count = 0
    
    predictor = CustomVisionPredictionClient(prediction_key, endpoint=base_url)
    
    for testfile in testfiles:
        predicts = {}
        with open(testfile, mode='rb') as f:
            results = predictor.classify_image(project_ID, publish_iteration_name, f.read())
            
        for prediction in results.predictions:
            predicts[prediction.tag_name] = prediction.probability
            
        prediction_result = max(predicts, key=predicts.get)
        
        if fishname == prediction_result:
            true_count += 1
            
    accuracy = (true_count / data_count) * 100
    print('正解率:' + str(accuracy) + '%')
```

```python
root_dir = 'fishimages/'
fishnames = ['アイゴ', 'オニカサゴ','カサゴ', 'カワハギ', 'キュウセンベラ',
             'クサフグ', 'ソウシハギ', 'マハゼ', 'マアジ', 'マイワシ', 'ミノカサゴ', 'メジナ', 'メバル']

for fishname in fishnames:
    print('****' + fishname + '****')
    testfiles = glob.glob(root_dir + fishname + '/test/*')
    testModel(testfiles, fishname)
```
